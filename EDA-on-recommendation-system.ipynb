```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
catalog = pd.read_csv('/kaggle/input/recommend-dataset/AI Recomendation Data Structure - Product Catalog - Product Catalog.csv')
category = pd.read_csv('/kaggle/input/recommend-dataset/AI Recomendation Data Structure - Product Category - Product Category.csv')
customers = pd.read_csv('/kaggle/input/recommend-dataset/AI Recommendation Data Structure - Customer.csv')
sales = pd.read_csv('/kaggle/input/recommend-dataset/AI Recommendation Data Structure - Sales.csv')
```

```python
# Merge datasets
product_data = pd.merge(catalog, category, on='product_id', how='inner')
sales_data = pd.merge(sales, customers, on='customer_id', how='inner')
full_data = pd.merge(sales_data, product_data, on='product_id', how='inner')
```

```python
# 1. Inspect the Data
print("Product Data Info:")
print(product_data.info())
```

```python
print("\nSales Data Info:")
print(sales_data.info())
```

```python
print("\nFull Data Info:")
print(full_data.info())
```

```python
# 2. Handle Missing Values
print("\nMissing values in Full Data:")
print(full_data.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap to visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(full_data.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Fill missing city values with 'Unknown'
full_data['city'] = full_data['city'].fillna('Unknown')
```

```python
# Drop the unnecessary columns and update `full_data`
full_data = full_data.drop(columns=['_id_x', '_id_y'])

# Check the updated column names to confirm
print("Updated Columns:")
print(full_data.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap to visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(full_data.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
print("\nMissing values in Full Data:")
print(full_data.isnull().sum())
```

```python
from datetime import datetime

# Ensure `dob` is a datetime type
full_data['dob'] = pd.to_datetime(full_data['dob'], errors='coerce')

# Calculate age
current_year = datetime.now().year
full_data['age'] = full_data['dob'].apply(lambda x: current_year - x.year if pd.notnull(x) else None)
```

```python
median_age = full_data['age'].median()
full_data['age'] = full_data['age'].fillna(median_age)
```

```python
full_data['product_description'] = full_data['product_description'].fillna('No description available')
```

```python
print(full_data['gender'].value_counts())
```

```python
full_data = full_data.drop(columns=['dob'])
```

```python
full_data['gender'] = full_data['gender'].fillna('Unknown')
```

```python
print("\nRemaining Missing Values:")
print(full_data.isnull().sum())
```

```python
# 3. Descriptive Statistics
print("\nDescriptive Statistics:")
print(full_data.describe())
```

```python
# 4. Analyze Customer Demographics
sns.countplot(data=full_data, x='gender')
plt.title('Customer Gender Distribution')
plt.show()
```

```python
# Plot Customer Age Distribution
sns.histplot(full_data['age'], bins=20, kde=True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Product Purchase Trends
full_data['purchased_date_time'] = pd.to_datetime(full_data['purchased_date_time'])
full_data['month_year'] = full_data['purchased_date_time'].dt.to_period('M')

plt.figure(figsize=(12, 6))
sns.countplot(data=full_data, x='month_year', order=full_data['month_year'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Product Purchase Trends Over Time')
plt.show()
```

```python
# 6. Outlier Detection
plt.figure(figsize=(8, 6))
sns.boxplot(data=full_data, x='quantity')
plt.title('Quantity Outlier Analysis')
plt.show()
```

```python
# Remove outliers (optional, based on findings)
Q1 = full_data['quantity'].quantile(0.25)
Q3 = full_data['quantity'].quantile(0.75)
IQR = Q3 - Q1
full_data = full_data[~((full_data['quantity'] < (Q1 - 1.5 * IQR)) | (full_data['quantity'] > (Q3 + 1.5 * IQR)))]

print("Data after removing outliers:")
print(full_data.describe())
```

```python
# Remove rows where age is less than 0
full_data = full_data[full_data['age'] >= 0]
```

```python
print("Data after removing age < 0:")
print(full_data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap to visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(full_data.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```
